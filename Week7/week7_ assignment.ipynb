# Handle missing values 

```python
from pyspark.sql.functions import col, sum, when

missing = df.select([
    sum(when(col(c).isNull(),1).otherwise(0)).alias(c)
    for c in df.columns
])

display(missing)
```

```python
df_clean=df.dropna()
```

# Basic operations

```python
from pyspark.sql.functions import col

df_clean = df.dropna() \
             .dropDuplicates() \
             .withColumn("Sales", col("Sales").cast("double")) \
             .withColumn("Quantity", col("Quantity").cast("int")) \
             .withColumn("Discount", col("Discount").cast("double"))
```

```python
display(df_clean.filter(col("Sales") > 500).limit(5))
```

```python
# Select required columns

selected_df = df_clean.select(
    "Order ID",
    "Customer Name",
    "Category",
    "Sub-Category",
    "Sales",
    "Quantity",
    "Profit"
)

display(selected_df.limit(5))
```

```python
# remove duplicate rows 
before = df_clean.count()
df_clean = df_clean.dropDuplicates()
after = df_clean.count()
print(f"Rows before removing duplicates:{before}")
print(f"Rows after removing duplicates:{after}")
print(f"Duplicate rows removed: {before - after}")
```

# No duplicates found!

```python
from pyspark.sql.functions import col

df_clean = df_clean.withColumn(
    "Unit Price",
    col("Sales") / col("Quantity")
)

df_clean = df_clean.withColumn(
    "total_amount",
    col("Unit Price") * col("Quantity")
)

display(
    df_clean.select(
        "Sales",
        "Quantity",
        "Unit Price",
        "total_amount"
    ).limit(10)
)
     
```

```python

# Rename columns for Delta compatibility

df_delta = df_clean.toDF(
    "Row_ID",
    "Order_ID",
    "Order_Date",
    "Ship_Date",
    "Ship_Mode",
    "Customer_ID",
    "Customer_Name",
    "Segment",
    "Country",
    "City",
    "State",
    "Postal_Code",
    "Region",
    "Product_ID",
    "Category",
    "Sub_Category",
    "Product_Name",
    "Sales",
    "Quantity",
    "Discount",
    "Profit",
    "Unit_Price",
    "total_amount"
)

display(df_delta.limit(50))
     
```

```python
delta_path = "/Volumes/sample_data/default/source/superstore_delta"

df_delta.write \
    .format("delta") \
    .mode("overwrite") \
    .save(delta_path)

print("Delta table created successfully!")
     
```
