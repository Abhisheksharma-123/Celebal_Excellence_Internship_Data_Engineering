# Import Libraries

```python
import pandas as pd
import numpy as np
import random
import os

from faker import Faker
```

# Create Faker Object

```python
fake = Faker()
```

# Create Folder Automatically

```python
os.makedirs("data/raw", exist_ok=True)
```

# Generate Customers Dataset

```python
customers = []

customer_types = ["REGULAR", "PREMIUM", "VIP"]

for i in range(1,501):

    email = fake.email()

    customers.append({

        "customer_id": i,

        "customer_name": fake.name(),

        "email": email,

        "registration_date": fake.date_between(start_date='-3y', end_date='today'),

        "customer_type": random.choice(customer_types)

    })

customers = pd.DataFrame(customers)

customers.head()
```

# Introduce Invalid Emails (2%)

```python
invalid_rows = customers.sample(frac=0.02).index

for i in invalid_rows:

    if random.choice([True, False]):

        customers.loc[i,"email"] = "abcgmail.com"

    else:

        customers.loc[i,"email"] = "xyz@yahoo"

customers.head()
```

# Save Customers CSV

```python
customers.to_csv("data/raw/customers.csv",index=False)

print("customers.csv saved successfully")
```

# Generate Products Dataset

```python
categories = {

    "Electronics":["Laptop","Mobile","Camera","Monitor","Keyboard"],

    "Clothing":["Shirt","Jeans","Jacket","Shoes","T-Shirt"],

    "Home":["Chair","Table","Sofa","Lamp","Fan"],

    "Books":["Python","SQL","Java","Spark","AI"]

}

products = []

for i in range(1,501):

    category = random.choice(list(categories.keys()))

    product = random.choice(categories[category])

    products.append({

        "product_id": i,

        "product_name": product,

        "category": category,

        "subcategory": product,

        "cost_price": round(random.uniform(100,10000),2)

    })

products = pd.DataFrame(products)

products.head()
```

# Introduce Product Name Errors

```python
random_rows = products.sample(frac=0.10).index

for i in random_rows:

    name = products.loc[i,"product_name"]

    style = random.randint(1,4)

    if style==1:

        products.loc[i,"product_name"] = " "+name+" "

    elif style==2:

        products.loc[i,"product_name"] = name.upper()

    elif style==3:

        products.loc[i,"product_name"] = name.lower()

    else:

        products.loc[i,"product_name"] = "  "+name.upper()

products.head()
```

# Save Products CSV

```python
products.to_csv("data/raw/products.csv",index=False)

print("products.csv saved successfully")
```

# Generate Orders Dataset

```python
statuses = [

    "PLACED",

    "SHIPPED",

    "DELIVERED",

    "CANCELLED",

    "RETURNED"

]

regions = [

    "North",

    "South",

    "East",

    "West"

]

orders = []

for i in range(1,1001):

    orders.append({

        "order_id": i,

        "customer_id": random.randint(1,500),

        "order_date": fake.date_time_between(start_date='-2y',end_date='now'),

        "status": random.choice(statuses),

        "region_code": random.choice(regions)

    })

orders = pd.DataFrame(orders)

orders.head()
```

# Add NULL Customer IDs (5%)

```python
null_rows = orders.sample(frac=0.05).index

orders.loc[null_rows,"customer_id"] = np.nan
```

# Wrong Date Format

```python
wrong_dates = orders.sample(frac=0.05).index

for i in wrong_dates:

    orders.loc[i,"order_date"] = pd.to_datetime(
        orders.loc[i,"order_date"]
    ).strftime("%d-%m-%Y")
```

# Save Orders CSV

```python
orders.to_csv("data/raw/orders.csv",index=False)

print("orders.csv saved successfully")
```

# Generate Order Items

```python
order_items = []

for i in range(1,3001):

    order_items.append({

        "item_id": i,

        "order_id": random.randint(1,1000),

        "product_id": random.randint(1,500),

        "quantity": random.randint(1,5),

        "unit_price": round(random.uniform(200,15000),2),

        "discount_percent": random.randint(0,50)

    })

order_items = pd.DataFrame(order_items)

order_items.head()
```

# Add Negative Quantities

```python
negative_rows = order_items.sample(frac=0.03).index

order_items.loc[negative_rows,"quantity"] *= -1
```

# Add Invalid Order IDs

```python
invalid_rows = order_items.sample(2).index

order_items.loc[invalid_rows,"order_id"] = [2001,2002]
```

# Save Order Items CSV

```python
order_items.to_csv("data/raw/order_items.csv",index=False)

print("order_items.csv saved successfully")
```
