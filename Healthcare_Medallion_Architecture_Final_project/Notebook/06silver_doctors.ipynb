```python
from pyspark.sql import functions as F
     
```

```python
doctors_df = (
    spark.read
    .format("delta")
    .load(f"{bronze_path}doctors")
)
display(doctors_df.limit(30))
```

```python
doctors_df.printSchema()

print("Total Records :", doctors_df.count())

display(doctors_df.describe())
```

# Remove Duplicates

```python
doctors_df = doctors_df.dropDuplicates(["doctor_id"])

print("After removing duplicates :", doctors_df.count())
```

# Hundle Null values

```python
doctors_df = doctors_df.na.drop(
    subset=["doctor_id"]
)
```

```python
doctors_df = doctors_df.fillna({
    "specialization": "GENERAL",
    "hospital_branch": "UNKNOWN"
})
```

```python
doctors_df = (
    doctors_df
    .withColumn("first_name", F.initcap(F.trim("first_name")))
    .withColumn("last_name", F.initcap(F.trim("last_name")))
    .withColumn("specialization", F.upper(F.trim("specialization")))
    .withColumn("hospital_branch", F.upper(F.trim("hospital_branch")))
    .withColumn("email", F.lower(F.trim("email")))
)
```

# Business Validations

```python
doctors_df = doctors_df.filter(
    F.col("years_experience") >= 0
)
```

```python
doctors_df = doctors_df.filter(
    F.col("email").rlike(r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$")
)
```

```python
doctors_df = doctors_df.withColumn(
    "doctor_name",
    F.concat_ws(" ", F.col("first_name"), F.col("last_name"))
)
```

```python
doctors_df = doctors_df.withColumn(
    "experience_level",
    F.when(F.col("years_experience") < 5, "JUNIOR")
     .when(F.col("years_experience") < 10, "MID")
     .otherwise("SENIOR")
)
```

```python
doctors_df = (
    doctors_df
    .withColumn("_dq_passed", F.lit(True))
    .withColumn("_silver_load_timestamp", F.current_timestamp())
    .withColumn("_silver_batch_id", F.lit(batch_id))
    .withColumn("_bronze_batch_id", F.col("_batch_id"))
    .withColumn("_is_current", F.lit(True))
    .withColumn("_record_version", F.lit(1))
)
```

```python
doctors_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(f"{silver_path}doctors")
```
