# Remove Duplicates

```python
treatments_df = treatments_df.dropDuplicates(["treatment_id"])
```

```python
print("After removing duplicates :", treatments_df.count())
```

# Remove Invalid Records

```python
treatments_df = treatments_df.na.drop(
    subset=[
        "treatment_id",
        "appointment_id"
    ]
)
```

```python
treatments_df = (

    treatments_df

    .withColumn(
        "treatment_type",
        F.upper(F.trim("treatment_type"))
    )

    .withColumn(
        "description",
        F.trim("description")
    )

)
```

```python
treatments_df = treatments_df.filter(
    F.col("cost") >= 0
)
```

```python
treatments_df = treatments_df.filter(
    F.col("treatment_date").isNotNull()
)
```

```python
treatments_df = treatments_df.join(

    appointments_df.select("appointment_id"),

    on="appointment_id",

    how="inner"

)
```

```python
treatments_df = treatments_df.withColumn(

    "high_cost_treatment",

    F.when(
        F.col("cost") >= 10000,
        1
    ).otherwise(0)

)
```

```python

treatments_df = treatments_df.withColumn(

    "treatment_category",

    F.when(
        F.col("cost") >= 5000,
        "PREMIUM"
    ).otherwise("STANDARD")

)
```

```python
treatments_df = (

    treatments_df

    .withColumn("_dq_passed", F.lit(True))

    .withColumn(
        "_silver_load_timestamp",
        F.current_timestamp()
    )

    .withColumn(
        "_silver_batch_id",
        F.lit(batch_id)
    )

    .withColumn(
        "_bronze_batch_id",
        F.col("_batch_id")
    )

    .withColumn(
        "_is_current",
        F.lit(True)
    )

    .withColumn(
        "_record_version",
        F.lit(1)
    )

)
```
