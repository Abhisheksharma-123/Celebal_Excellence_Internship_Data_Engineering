```python
patients_df = (
    spark.read
    .format("delta")
    .load(f"{silver_path}patients")
)
```

```python
appointments_df.printSchema()

print("Rows :", appointments_df.count())

display(appointments_df.describe())
```

# Remove Duplicate

```python

appointments_df = appointments_df.dropDuplicates(["appointment_id"])
     
```

```python
print("After removing duplicates :", appointments_df.count())
```

# Remove Invalid Records

```python
appointments_df = appointments_df.na.drop(
    subset=[
        "appointment_id",
        "patient_id",
        "doctor_id"
    ]
)
```

# Standardize Text Columns

```python
appointments_df = (
    appointments_df
    .withColumn(
        "reason_for_visit",
        F.trim("reason_for_visit")
    )
    .withColumn(
        "status",
        F.upper(F.trim("status"))
    )
)
```

```python
appointments_df = appointments_df.filter(
    F.col("appointment_date").isNotNull()
)
```

```python
appointments_df = appointments_df.join(

    patients_df.select("patient_id"),

    on="patient_id",

    how="inner"

)
```

```python
appointments_df = appointments_df.withColumn(
    "no_show_flag",
    F.when(
        F.col("status") == "NO SHOW",
        1
    ).otherwise(0)
)
     
```

```python
appointments_df = appointments_df.withColumn(

    "is_weekend",

    F.when(

        F.dayofweek("appointment_date").isin([1,7]),

        1

    ).otherwise(0)

)
```

```python
appointments_df = (

    appointments_df

    .withColumn(
        "_dq_passed",
        F.lit(True)
    )

    .withColumn(
        "_silver_load_timestamp",
        F.current_timestamp()
    )

    .withColumn(
        "_silver_batch_id",
        F.lit(batch_id)
    )

    .withColumn(
        "_bronze_batch_id",
        F.col("_batch_id")
    )

    .withColumn(
        "_is_current",
        F.lit(True)
    )

    .withColumn(
        "_record_version",
        F.lit(1)
    )

)
```

```python
appointments_df.write \
.format("delta") \
.mode("overwrite") \
.save(f"{silver_path}appointments")
```
